```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from safetensors.torch import load_file as safe_load_file
import torch
```

```python
# Load pre-trained BERT tokenizer
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
import os
# Load the model configuration and weights from safetensors
model_path = "sentiment model"  # Update this path to your model directory
config_path = os.path.join(model_path, "config.json")
model_weights_path = os.path.join(model_path, "model.safetensors")

model = AutoModelForSequenceClassification.from_pretrained(model_path)
model.eval()

```

```python
# Function to analyze sentiment with three classes: Positive, Neutral, Negative
def analyze_sentiment(text):
    inputs = bert_tokenizer(text, return_tensors="pt", max_length=512, truncation=True)
    outputs = model(**inputs)
    predicted_class = torch.argmax(outputs.logits, dim=1).item()

    if predicted_class == 0:
        return "Negative"
    elif predicted_class == 1:
        return "Neutral"
    elif predicted_class == 2:
        return "Positive"
    else:
        return "Unknown"  # In case the model outputs an unexpected class

```

```python
# Example usage
review_text = input('Enter your review: ')
sentiment = analyze_sentiment(review_text)
print(f"Sentiment: {sentiment}")
```
