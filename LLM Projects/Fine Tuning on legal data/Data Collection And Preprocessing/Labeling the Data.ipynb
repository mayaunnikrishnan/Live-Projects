```python
import pandas as pd
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('legaljanuary.csv')
```

```python
df.info()
```

```python
key_to_label = {
    "ACCEPTED": "ACCEPTED",
    "Accepted": "ACCEPTED",
    "REJECTED": "REJECTED",
    "Rejected": "REJECTED",
    "RANK1": "RANK 1",
    "Rank1": "RANK 1",
    "Rank2": "RANK 2",
    "Rank3": "RANK 3",
    "Rank4": "RANK 4",
    "Rank5": "RANK 5",
    "Rank6": "RANK 6",
    "Rank7": "RANK 7",
    "Rank8": "RANK 8",
    "Rank9": "RANK 9"
}
```

```python
list1=[]
for i in range (len(df['fact'])) :
    for key,value in key_to_label.items():
        if key.lower() in df['fact'][i].lower():
            id=key
    list1.append(id)
df['Result']=list1
```

```python
df['Result'].value_counts()
```

```python
# Find null values in the "Result" column
null_values = df["Result"].isnull()

# Display the rows where the "Result" column has null values
df[null_values]
```

```python
label=[]
for i in range (len(df['Result'])):
    if df['Result'][i]=='Rejected':
        label.append(0)
    else:
        label.append(1)
df['Label']=label
```
